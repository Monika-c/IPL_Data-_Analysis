```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\ipl_matches_100.csv")
df.head()
df.info()
df.describe()
```

```python
team_wins = df['match_winner'].value_counts()
team_wins
```

```python
plt.figure()
team_wins.plot(kind='bar')
plt.title("Total Wins by Team")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['toss_match_win'] = df['toss_winner'] == df['match_winner']
toss_impact = df['toss_match_win'].value_counts()
toss_impact
```

```python
toss_win_percentage = df['toss_match_win'].mean() * 100
print("Toss impact percentage:", toss_win_percentage)
```

```python
plt.figure()
toss_impact.plot(kind='pie', autopct='%1.1f%%')
plt.title("Toss Winner Also Won Match")
plt.ylabel("")
plt.show()
```

```python
season_wins = df.groupby('season')['match_winner'].count()

plt.figure()
season_wins.plot()
plt.title("Matches Played per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()
```

```python
venue_wins = df['venue'].value_counts().head(10)

plt.figure()
venue_wins.plot(kind='bar')
plt.title("Top 10 Venues by Matches")
plt.xticks(rotation=45)
plt.show()
```
